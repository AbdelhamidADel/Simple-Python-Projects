```python
import cv2
from matplotlib import pyplot as plt
import yaml
print("           Available Books In Library Now         ")
print("--------------------------------------------------")
print("If you are interested, hurry now to the library!")
fig = plt.figure(figsize=(10, 7))

rows = 1
columns = 5
Image1 = cv2.imread('In Search of Lost Time.jpg')
Image2 = cv2.imread('ulysses.jpg')
Image3 = cv2.imread('DonQuixoteCover.jpg')
Image4 = cv2.imread('One Hundred Years of Solitude.jpg')
Image5 = cv2.imread('Moby Dick.jpg')

fig.add_subplot(rows, columns, 1);
plt.imshow(Image1);
plt.axis('off')
plt.title("In Search of Lost Time",fontsize=7);
  
fig.add_subplot(rows, columns, 2);
plt.imshow(Image2);
plt.axis('off')
plt.title("Ulysses",fontsize=7);
  
fig.add_subplot(rows, columns, 3);
plt.imshow(Image3);
plt.axis('off')
plt.title("Don Quixote",fontsize=7);
  
fig.add_subplot(rows, columns, 4);
plt.imshow(Image4);
plt.axis('off')
plt.title("One Hundred Years of Solitude",fontsize=7);

fig.add_subplot(rows, columns, 5);
plt.imshow(Image5);
plt.axis('off')
plt.title("Moby Dick",fontsize=7);


```

```python

class Library:
    
    def __init__(self, list, name):
        self.booksList = list
        self.name = name
        self.lendDict = {}
        
    def displayBooks(self):
        print(f"We have following books in our library{self.name}")
        print("--------------------------------------------------")
        for book in self.booksList:
            print(yaml.dump(book, sort_keys=True, default_flow_style=False))           
        
    def lendBook(self, user, book):
        if book not in self.lendDict.keys():
            self.lendDict.update({book:user})
            print("Lender-Book database has been updated. You can take the book now")
        else:
            print(f"Book is already being used by {self.lendDict[book]}")

    def addBook(self, book):
        self.booksList.append(book)
        print("Book has been added to the book list")

    def returnBook(self, book):
        self.lendDict.pop(book)

if __name__ == '__main__':
    public_library = Library([{"-In Search of Lost Time":"by Marcel Proust"},{"-Ulysses":"by James Joyce"},
        {"-Don Quixote":"by Miguel de Cervantes"},{"-One Hundred Years of Solitude":"by Gabriel Garcia Marquez"},
                      {"-Moby Dick":"by Herman Melville"}],"Sea of Books")

    while(True):
        print(f"Welcome to the {public_library.name} library"+"\n"+"Enter your choice to continue"+"\n")
        print("1. Display Books")
        print("2. Lend a Book")
        print("3. Add a Book")
        print("4. Return a Book")
        print("--------------------------------------------------")
        print("         Hint        ")
        print("---------------------")
        print("Press c For Continue")
        print("Press q For Close Program")
        user_choice = input()
        if user_choice not in ['1','2','3','4',"q","c"]:
            print("Please enter a valid option !")
            continue
        elif user_choice == "q":
            break

        elif user_choice == "c":
            continue        

        else:
            user_choice = int(user_choice)


        if user_choice == 1:
            public_library.displayBooks()
       
        elif user_choice == 2:
            book = input("Enter the name of the book you want to lend:")
            user = input("Enter your name")
            public_library.lendBook(user, book)

        elif user_choice == 3:
            book = input("Enter the name of the book you want to add:")
            public_library.addBook(book)

        elif user_choice == 4:
            book = input("Enter the name of the book you want to return:")
            public_library.returnBook(book)

        else:
            print("Not a valid option !")
        
       

```
